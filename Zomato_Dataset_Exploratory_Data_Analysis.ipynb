```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
#we combine the country_code like df_country_code and df
final_df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

```python
#Pie chart-- for top 3 counties that uses zomato
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

observations: Zomato maximum records are from India
After than USA and then United Kingdoms

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

#observation
1. when rating is between 4.5 to 4.9 --> Excellent
2. when ratings are between 4.0 to 3.9  --> very good
3. when rating is between 3.5 to 3.9 --> Good
4. when rating is between 2.5 to 3.4 --> Average
5. when rating is between 1.8 to 2.4 --> Poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (11,3)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,hue='Rating color',palette=['blue','Red','orange','yellow','green','green'])
```

Observation:

1. No rated count is very high
2. Max number of ratings are between 2.5 to 3.4

```python
#count plot -plot using categorical values how many records in each color for each ratings
sns.countplot(x='Rating color',data=ratings,palette=['blue','Red','orange','yellow','green','green'])
```
